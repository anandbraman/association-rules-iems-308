# Preprocessing outline

**1)** random sample of 5 stores
<br>
**2)** filter for stype == "P"
<br>
**3)** one hot encode sku
<br>
**4)** group by store, register, trannum, seq, saledate and sum
<br>
**5)** iterate over cols to make sure that no sum > 1
<br>
**6)** export final data frame to csv and analyze in `association_rules.ipynb`

```python
random.seed(308)
store_samp = random.sample(strinfo.store.tolist(), 5)
print(store_samp)
```

```python
chunk_lst = []
for chunk in trnsact_chunk:
    chunk_filt = chunk[(chunk.store.isin(store_samp)) & (chunk.stype == "P")]
    chunk_lst.append(chunk_filt)
```

```python
trnsact_df = pd.concat(chunk_lst)
trnsact_df.shape
```

```python
# dropping stype
trnsact_df = trnsact_df.drop('stype', axis=1)
```

```python
# filtering out some skus that do not occur frequently
sku_counts = trnsact_df.sku.value_counts()
# not useful to try to find association rules for things that only have a few obs
print(len(sku_counts))
print(0.0001 * len(trnsact_df.index))
```

```python
cutoff = 0.0001 * len(trnsact_df.index)
sku_count_filt = sku_counts[sku_counts >= cutoff]
```

```python
final_skus = sku_count_filt.index.tolist()
trnsact_df_filt = trnsact_df[trnsact_df.sku.isin(final_skus)]
```

```python
# dummy code sku
sku_dummy = pd.get_dummies(trnsact_df_filt['sku'], prefix="sku")
```

```python
sku_dummy.head()
```

```python
trnsact_df_dummy = pd.concat([trnsact_df_filt, sku_dummy], axis=1)
```

```python
trnsact_df_dummy.shape
```

```python
trnsact_df_dummy = trnsact_df_dummy.drop('sku', axis=1)
```

```python
trnsact_df_dummy.head()
```

```python
trnsact_df_grouped = trnsact_df_dummy.groupby(['store', 'register', 'trannum', 'saledate', 'seq']).sum()
```

```python
trnsact_df_grouped.head(15)
```

```python
trnsact_df_grouped.shape
```

```python
repeat_lst = []
for col in trnsact_df_grouped.columns:
    repeat = trnsact_df_grouped[trnsact_df_grouped[col] > 1]
    repeat_lst.append(repeat)
```

```python
# making max value 1 in every column because this is how assoc rule algo works

for col in trnsact_df_grouped.columns:
    trnsact_df_grouped[col] = np.where(trnsact_df_grouped[col] > 1, 1, trnsact_df_grouped[col])
```

```python
repeat_lst = []
for col in trnsact_df_grouped.columns:
    repeat = trnsact_df_grouped[trnsact_df_grouped[col] > 1]
    repeat_lst.append(repeat)
```

```python
pd.concat(repeat_lst)
# heck yeah it worked!
```
